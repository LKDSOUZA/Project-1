```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np

crime_path = "crimedata1.csv"
crime_data = pd.read_csv(crime_path)
crime_data.head()
```

```python
# crime_data_covid = crime_data["Year"] >= "2018"

# crime_data_covid=(crime_data[crime_data["Year"]>=2018])
crime_data_21 = crime_data.loc[crime_data["Year"] == 2021]
crime_data_21
```

```python
statistics_raw_data=crime_data_21.copy()

victim_report_mean=statistics_raw_data.groupby(['Offence Division'])["Rate per 100,000 population"].mean()
victim_report_median=statistics_raw_data.groupby(['Offence Division'])["Rate per 100,000 population].median()
victim_report_variance=statistics_raw_data.groupby(['Offence Division'])["Rate per 100,000 population"].var()
victim_report_sd=statistics_raw_data.groupby(['Offence Division'])["Rate per 100,000 population"].std()
victim_report_SEM=statistics_raw_data.groupby(['Offence Division'])["Rate per 100,000 population"].sem()

statistics_summary = pd.DataFrame({"Mean victim report": victim_report_mean
                                  "Median victim report": victim_report_median,
                                  "victim report Variance": victim_report_variance,
                                  "victim report Std. Dev.": victim_report_sd,
                                  "victim report Std. Err." : victim_report_SEM,
                                 
                                 })
statistics_summary.head(10)
```

```python
bar_plot = crime_data_21['Offence Subdivision'].value_counts().plot.bar(figsize=(5,4))

bar_plot.set_xlabel("Offence Subdivision")
bar_plot.set_ylabel("Number")

plt.tight_layout()
```
