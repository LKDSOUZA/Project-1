```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np


crime_path = "crimedata1.csv"
crime_data = pd.read_csv(crime_path)
crime_data.head()
```

```python
# Getting data for 2021
crime_data_21 = crime_data.loc[crime_data["Year"] == 2021]
crime_data_21["Victim Reports"].str.replace(",","").astype(int)
crime_data_21["Victim Reports"] = crime_data_21["Victim Reports"].str.replace(",","").astype(int)

# Getting data for 2020
crime_data_20 = crime_data.loc[crime_data["Year"] == 2020]
crime_data_20["Victim Reports"].str.replace(",","").astype(int)
crime_data_20["Victim Reports"] = crime_data_20["Victim Reports"].str.replace(",","").astype(int)

# Getting data for 2019
crime_data_19 = crime_data.loc[crime_data["Year"] == 2019]
crime_data_19["Victim Reports"].str.replace(",","").astype(int)
crime_data_19["Victim Reports"] = crime_data_19["Victim Reports"].str.replace(",","").astype(int)

# Getting data for 2018
crime_data_18 = crime_data.loc[crime_data["Year"] == 2018]
crime_data_18["Victim Reports"].str.replace(",","").astype(int)
crime_data_18["Victim Reports"] = crime_data_18["Victim Reports"].str.replace(",","").astype(int)
```

```python
statistics_raw_data=crime_data_21.copy()

victim_report_mean_21=statistics_raw_data.groupby(['Offence Subdivision'])["Victim Reports"].mean()
victim_report_median_21=statistics_raw_data.groupby(['Offence Subdivision'])["Victim Reports"].median()
victim_report_sum_21=statistics_raw_data.groupby(['Offence Subdivision'])["Victim Reports"].sum()


statistics_summary_21 = pd.DataFrame({"Mean victim report": victim_report_mean_21,
                                  "Median victim report": victim_report_median_21,
                                  "Total victim report": victim_report_sum_21,
                                  
                                 })
statistics_summary_21.head(10)
```

```python
statistics_raw_data_20=crime_data_20.copy()

victim_report_mean_20=statistics_raw_data_20.groupby(['Offence Subdivision'])["Victim Reports"].mean()
victim_report_median_20=statistics_raw_data_20.groupby(['Offence Subdivision'])["Victim Reports"].median()
victim_report_sum_20=statistics_raw_data_20.groupby(['Offence Subdivision'])["Victim Reports"].sum()

statistics_summary_20 = pd.DataFrame({"Mean victim report": victim_report_mean_20,
                                  "Median victim report": victim_report_median_20,
                                  "Total victim report": victim_report_sum_20
                                  
                                 
                                 })
statistics_summary_20.head(10)
```

```python
statistics_raw_data_19=crime_data_19.copy()

victim_report_mean_19=statistics_raw_data_19.groupby(['Offence Subdivision'])["Victim Reports"].mean()
victim_report_median_19=statistics_raw_data_19.groupby(['Offence Subdivision'])["Victim Reports"].median()
victim_report_sum_19=statistics_raw_data_19.groupby(['Offence Subdivision'])["Victim Reports"].sum()


statistics_summary_19 = pd.DataFrame({"Mean victim report": victim_report_mean_19,
                                  "Median victim report": victim_report_median_19,
                                  "Total victim report": victim_report_sum_19,
                                
                                 })
statistics_summary_19.head(10)
```

```python
statistics_raw_data_18=crime_data_18.copy()

victim_report_mean_18=statistics_raw_data_18.groupby(['Offence Subdivision'])["Victim Reports"].mean()
victim_report_median_18=statistics_raw_data_18.groupby(['Offence Subdivision'])["Victim Reports"].median()
victim_report_sum_18=statistics_raw_data_18.groupby(['Offence Subdivision'])["Victim Reports"].sum()


statistics_summary_18 = pd.DataFrame({"Mean victim report": victim_report_mean_18,
                                  "Median victim report": victim_report_median_18,
                                  "Total victim report": victim_report_sum_18,
                                 
                                 
                                 })
statistics_summary_18.head(10)
```

```python
victim_report_year= pd.DataFrame({"2018 Victim Report":victim_report_sum_18,
                             "2019 Victim Report":victim_report_sum_19,
                            "2020 Victim Report":victim_report_sum_20,
                            "2021 Victim Report":victim_report_sum_21,  
                            })
victim_report_year
```

```python
df = pd.DataFrame(dict(graph=[
                              'A10 Homicide and related offences', 
                              'A20 Assault and related offences', 
                              'A30 Sexual offences',
                              'A40 Abduction and related offences',
                              'A50 Robbery',
                              'A60 Blackmail and extortion',
                              'A70 Stalking, harassment and threatening behaviour',
                              'A80 Dangerous and negligent acts endangering people',
                              'B10 Arson',
                              'B20 Property damage',
                              'B30 Burglary/Break and enter',
                              'B40 Theft',
                              'B50 Deception',
                              'C Other offences'], 
                              n=list(victim_report_year['2018 Victim Report']),
                              m=list(victim_report_year['2019 Victim Report']),
                              o=list(victim_report_year['2020 Victim Report']),
                              p=list(victim_report_year['2021 Victim Report'])))

ind = np.arange(len(df))
width = 0.2



fig, ax = plt.subplots()



ax.barh(ind, df.p, width, color = 'lightskyblue',label ='2021')
ax.barh(ind + width, df.o,width,color = 'orange',label ='2020')
ax.barh(ind + width*2, df.m,width,color = 'green',label ='2019')
ax.barh(ind + width*3, df.n,width,color = 'lightcoral',label ='2018')



ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width -1, len(df)])

ax.legend()



plt.savefig("./victim_report.png")

plt.show()
```
